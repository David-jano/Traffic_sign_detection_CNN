```python
import cv2
from ultralytics import YOLO
import pyttsx3  # Text-to-speech library

# Initialize the text-to-speech engine
engine = pyttsx3.init()

# Set the voice to "Amlie" explicitly
voices = engine.getProperty('voices')
for voice in voices:
    if "Samantha" in voice.name:  # Replace "Amlie" with the desired voice name
        engine.setProperty('voice', voice.id)
        break

# Load the trained YOLO model
model = YOLO("best.pt")  # Replace with the actual path to your trained model

# Initialize webcam
cap = cv2.VideoCapture(0)

# Set to store detected classes to avoid repeating the same class
detected_classes = set()

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        print("Failed to grab frame")
        break
    
    # Resize frame to match training resolution
    frame = cv2.resize(frame, (640, 640))

    # Perform object detection with a lower confidence threshold
    results = model(frame, conf=0.1)  # Run inference on the frame
    
    # Debug: Print model output
    print(results[0].boxes)  # Print bounding boxes, class IDs, and confidence scores

    # Get the first result
    result = results[0]

    # Get detected class names
    class_names = result.names  # Dictionary mapping class IDs to names
    detected_classes_current = set()  # Store classes detected in the current frame

    for box in result.boxes:
        class_id = int(box.cls)  # Get class ID
        class_name = class_names[class_id]  # Get class name
        detected_classes_current.add(class_name)  # Add to current frame's detected classes

    # Announce new classes detected in this frame
    for class_name in detected_classes_current:
        if class_name not in detected_classes:
            print(f"Detected: {class_name}")
            engine.say(f"Detected {class_name}")  # Add text to the speech queue
            detected_classes.add(class_name)  # Add to the set of detected classes

    # Speak the queued text
    engine.runAndWait()

    # Clear the set of detected classes for the next frame
    detected_classes.clear()
    detected_classes.update(detected_classes_current)

    # Plot the detections on the frame
    frame = result.plot()  # 'plot()' replaces the old 'render()' function
    
    # Display the frame with detections
    cv2.imshow("Real-time Traffic Sign Detection system", frame)

    # Break the loop when 'q' is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the webcam and close OpenCV windows
cap.release()
cv2.destroyAllWindows()
```
