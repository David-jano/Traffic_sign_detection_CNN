```python
import cv2
import torch
import torchvision.transforms as transforms
import torchvision.models as models
import pyttsx3  # Text-to-speech library

# Initialize the text-to-speech engine
engine = pyttsx3.init()

# Set the voice to "Samantha"
voices = engine.getProperty('voices')
for voice in voices:
    if "Samantha" in voice.name:
        engine.setProperty('voice', voice.id)
        break

# Define number of output classes for traffic signs
num_classes = 43

# Define the model architecture
model = models.resnet18(pretrained=False)
model.fc = torch.nn.Linear(model.fc.in_features, num_classes)

# Load your trained weights
model.load_state_dict(torch.load("resnet_Traffic_sign.pt", map_location=torch.device('cpu')))
model.eval()

# Define image transforms
transform = transforms.Compose([
    transforms.ToPILImage(),
    transforms.Resize((224, 224)),  # Match ResNet input size
    transforms.ToTensor(),
])

# Define your class labels (you need to fill in the actual traffic sign labels here)
class_labels = [f"Sign {i}" for i in range(num_classes)]  # Replace with actual labels if available

# Initialize webcam
cap = cv2.VideoCapture(0)

# Set to store detected classes to avoid repeating the same class
detected_classes = set()

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        print("Failed to grab frame")
        break

    # Copy frame to draw on later
    draw_frame = frame.copy()

    # Preprocess the frame
    input_tensor = transform(frame).unsqueeze(0)

    # Perform classification
    with torch.no_grad():
        output = model(input_tensor)
        pred_class_id = torch.argmax(output, dim=1).item()
        class_name = class_labels[pred_class_id]

    # Announce if new class detected
    if class_name not in detected_classes:
        print(f"Detected: {class_name}")
        engine.say(f"Detected {class_name}")
        engine.runAndWait()
        detected_classes.add(class_name)

    # Draw text on the frame
    cv2.putText(draw_frame, f"Detected: {class_name}", (10, 30),
                cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)

    # Display the frame
    cv2.imshow("Traffic Sign Classifier", draw_frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release resources
cap.release()
cv2.destroyAllWindows()
```
